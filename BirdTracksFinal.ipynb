```python
def validation_split(C, ratio):
    # Shuffle de data
    np.random.shuffle(C)
    tot_length = len(C)
    # Selecteer een training set en validation set op basis van van de ratio
    training_length = int(len(C) * ratio)
    training_C = C[0:training_length]
    validation_C = C[training_length:tot_length]
    return training_C, validation_C
```

```python
rain_training, rain_validation = validation_split(rainarray, 0.7)
bird_training, bird_validation = validation_split(birdarray, 0.7)
```

```python
########SCORE#########
mbirds_score = np.mean(bird_training[:,[10]])
mrain_score = np.mean(rain_training[:,[10]])
sbirds_score = np.std(bird_training[:,[10]])
srain_score = np.std(rain_training[:,[10]])
```

```python
def PDF(x, m, s):
    PDF = 1/(np.sqrt(2*np.pi)*s) * (np.e**(-((x-m)**2)/(2*s**2)))
    return(PDF)

# Genereer x-waardes om een plot mogelijk te maken
x_values = np.linspace(0,1,100)
    
plt.plot(x_values, PDF(x_values, mbirds_score, sbirds_score), label='PDF birds')
plt.plot(x_values, PDF(x_values, mrain_score, srain_score), label='PDF rain')
plt.xlabel('score')
plt.ylabel('probability')
plt.legend()
plt.show()
```

```python
def posteriors(x, m, s):
    posterior = PDF(x, m, s) / (PDF(x, mbirds_score, sbirds_score)+PDF(x, mrain_score, srain_score))
    return posterior

# Maak een mean vector en een standard deviation vector
m_vector = [mbirds_score, mrain_score]
s_vector = [sbirds_score, srain_score]

x_values_post = np.linspace(0,1,100)

# Plot de posteriors
plt.plot(x_values_post, posteriors(x_values_post, m_vector[0], s_vector[0]), label='Birds')
plt.plot(x_values_post, posteriors(x_values_post, m_vector[1], s_vector[1]), label='Rain')
plt.xlabel('score')
plt.ylabel('probability')
plt.legend()
plt.show()
```

```python
#Standardize data
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

def pcafunc(df):
    pca = PCA(n_components=2)

    featureslist = []
    for col in df.columns:
        featureslist.append(col)
#     print(featureslist)

    # Separating out the features
    x = df.loc[:, featureslist].values
    x = pd.DataFrame(x).T

    #y = df.loc[:,['target']].values

    # Standardizing the features
    x = StandardScaler().fit_transform(x)
    principalComponents = pca.fit_transform(x)
    principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])
    return principalDf

Birdpca = pcafunc(Bird)
Rainpca = pcafunc(Rain)
```

```python
def classify(x, m, s):
    # Bereken alle posteriors gegeven datapunt x
    Cl0 = posteriors(x, m[0], s[0])
    Cl1 = posteriors(x, m[1], s[1])
    # Return de grootste waarde
    if Cl0 > Cl1:
        return(0)
    else:
        return(1)


def validate(val, expected, m, s):
    expected_list = []
    # Vind voor iedere petal length uit de validation set de verwachtte bloemsoort
    for element in val:
        expected_list.append(classify(element[10], m_vector, s_vector))
    teller, noemer = 0, 0
    # Check of de daadwerkelijke soort klopt met het verwachtte soort
    for element in expected_list:
        if element == expected:
            teller += 1
            noemer += 1
        else:
            noemer += 1
    # Bereken het percentage juiste verwachtingen
    percentage = round(teller/noemer * 100)
    return percentage

val_C0 = validate(bird_validation, 0, m_vector, s_vector)
val_C1 = validate(rain_validation, 1, m_vector, s_vector)

average_val = round((val_C0+val_C1)/2)

print('validation birds =', val_C0, '%')
print('validation rain =', val_C1, '%')
print('average validation =', average_val, '%')
```
