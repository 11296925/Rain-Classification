```python
%matplotlib inline

import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tracks = pd.read_csv("LnB_tracks_MarToNov18.csv")
tracks = tracks.sort_values(["id"])
tracks
```

```python
rain = pd.read_csv("LnB_raintrack_ids.csv")
rain = rain.sort_values(["id"])
rain_id = np.array(rain["id"])
```

```python
rain, norain = [], []
i = 220819
for track_id in np.array(tracks["id"]):
    if rain_id[i] == track_id:
        i+=1
        rain.append('1')
    else:
        norain.append('0')    
```

```python
rainbinarylist = []
i = 220819
for track_id in np.array(tracks["id"]):
    if rain_id[i] == track_id:
        i+=1
        rainbinarylist.append(1)
    else:
        rainbinarylist.append(0)
```

```python
df = tracks
df = df.drop(['classification_id'],axis = 1)

se = pd.Series(rainbinarylist)
df['rainbinary'] = se.values
df = df.drop(['timestamp_start','timestamp_end','tracktype'],axis = 1)
df
```

```python
df_rain = df.loc[df.rainbinary == 1]
#print(len(df_rain))
df_norain = df.loc[df.rainbinary == 0].sample(n = len(df_rain))
new_random_df = pd.concat([df_rain, df_norain])
#print(df_rain)
#print(df_norain)
new_random_df = new_random_df.reindex(np.random.permutation(new_random_df.index))   
    
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
def regenteller(data):
    raincounter = 0
    noraincounter = 0
    justrain = np.array(data)
    #print(justrain)
    for i in range(len(justrain)):
        if justrain[i] == 1:
            raincounter += 1
        else:
            noraincounter += 1
    return raincounter

```

```python
TreeX = df.values
TreeY = TreeX[:,-1]
TreeX = TreeX[:,0:8]
X_train, X_test, Y_train, Y_test = train_test_split(TreeX, TreeY, test_size=0.3)
```

```python
#####FULL DATASET####
rf = RandomForestClassifier(n_estimators = 10)
rf.fit(X_train, Y_train)
y_predictforest = rf.predict(X_test)
print('Test accuracy random forest:',accuracy_score(Y_test,y_predictforest,)*100,'%')
print('Regen in testdata: ', regenteller(Y_test))
print('Gevonden regen: ', regenteller(y_predictforest))
print('Percentage juist gevonden regen:', regenteller(y_predictforest)/regenteller(Y_test))
```

```python
rain_correct, false_positives, false_negatives, norain_correct = 0, 0, 0, 0
falsepos_list, falseneg_list = [], []

for i in range(len(Y_test)):
    if y_predictforest[i] == 1 and Y_test[i] == 1:
        rain_correct += 1
    elif y_predictforest[i] == 0 and Y_test[i] == 1:
        false_negatives += 1
        falseneg_list.append(X_test[i])
    elif y_predictforest[i] == 1 and Y_test[i] == 0:
        false_positives += 1
        falsepos_list.append(X_test[i])
    else:
        norain_correct += 1
print('rain correctly classifiedrain as rain:', rain_correct)
print('No rain correctly classifiedrain as no rain:', norain_correct)
print()
print('No rain classified as rain:', false_positives)
print('Rain classified as no rain:', false_negatives)
print()
print('Percentage correct gelabeld overall:', (rain_correct+norain_correct)/300000*100, '%')
```

```python
print('Percentage of tracks classified as rain currently labeled as no rain:', len(falsepos_list)/(len(Y_test)-regenteller(Y_test))*100,'%')
print('Percentage of tracks classified as no rain currently labeled as rain:', len(falseneg_list)/regenteller(Y_test)*100,'%')
```
